```python
import pandas as pd
```

```python
clg_placement = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/CollegePlacement.csv')
```

```python
clg_placement.head()
```

```python
clg_placement.info()
```

```python
clg_placement.describe()
```

```python
clg_placement.columns
```

```python
y = clg_placement['Placement']
x = clg_placement[['Python', 'DeepLearning', 'MachineLearning', 'BigData', 'CloudComputing']]
y = pd.DataFrame(y).replace({'Placement': {'Yes': 1, 'No': 0}})
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=1)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
